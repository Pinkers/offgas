```python
"""import the package and initialize the instance, 
you can run this cell any time before you want to begin collecting data"""

import offgas as ofgs
run = ofgs.Offgasmonitor('enter name here')
```

```python
"""this begins data collection! Run this cell after inoculation"""

run.begin_collecting_data()
```

```python
"""running this cell will refresh the plot with the data up to now"""

run.data.plot(secondary_y='temp')
```

```python
"""This command stops data collection and exports the dataframe to a CSV and the plot to a png
in the folder under the name used to initialize this run in offgas_data. Run it before tear-down"""

run.finish_collecting_data()
```
